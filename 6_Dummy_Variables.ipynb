```r
# --- Example 1: mtcars Dataset ---
# Load data
data("mtcars")

# Convert 'am' (Transmission) to factor: 0=Automatic, 1=Manual
mtcars$am <- factor(mtcars$am, levels = c(0, 1), labels = c("Automatic", "Manual"))

# Perform Regression
model <- lm(mpg ~ am, data = mtcars)
summary(model)

# --- Example 2: iris Dataset ---
# Load data
data("iris")

# Convert Species to factor
iris$Species <- factor(iris$Species)

# Perform Regression with Dummy Variables
model_iris <- lm(Sepal.Length ~ Species + Sepal.Width + Petal.Length, data = iris)
summary(model_iris)
```
