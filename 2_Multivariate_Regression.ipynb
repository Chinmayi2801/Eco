```r
# File Required: mammalsize.xlsx
# Install and load packages
install.packages("MASS")
install.packages("readxl")
install.packages("lmtest")
library(readxl)
library(MASS)
library(lmtest)
```

```r

# 1. Import Data
mammal <- read_excel("mammalsize.xlsx")
names(mammal)
```

```r
# 2. Define Variables
y2 = mammal$"gestation period (days)"
x21 = mammal$"brain weight (gms)"
x22 = mammal$"body weight(kg)"
x23 = mammal$"av. Litter size"
```

```r
# 3. Fit Initial Model
model2 = lm(y2 ~ x21 + x22 + x23)
summary(model2)
```

```r
# 4. Diagnostic Plots (Check for Outliers)
plot(model2)
```

```r
# 5. Remove Outlier (Row 72) and Re-fit
mammal.1 = mammal[-c(72),]
```

```r
y2.1 = mammal.1$"gestation period (days)"
x21.1 = mammal.1$"brain weight (gms)"
x22.1 = mammal.1$"body weight(kg)"
x23.1 = mammal.1$"av. Litter size"
```

```r
model2.1 = lm(y2.1 ~ x21.1 + x22.1 + x23.1)
summary(model2.1)
plot(model2.1)
```

```r
# 6. Second Outlier Removal (If Row 77 is an outlier)
mammal.2 = mammal.1[-c(77),]
# Note: In Colab, ensure the index aligns with your dataframe
# Re-define variables and fit model2.2 (logic follows same pattern)
```

```r
# 7. Log Transformation (Remedial Measure)
y2new = log(y2)
model2.3 = lm(y2new ~ x21 + x22 + x23) # Using original x variables for demo
summary(model2.3)
plot(model2.3)
```

```r
# Check Assumptions
bptest(model2.3)
shapiro.test(y2new)
```
