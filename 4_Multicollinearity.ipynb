```r
# File Required: Multicoliratity_data.xlsx
# Install and load packages
install.packages("MASS")
install.packages("readxl")
install.packages("mctest")
install.packages("glmnet")
library(MASS)
library(readxl)
library(mctest)
library(glmnet)
```

```r
# 1. Import Data
gdp = read_excel("Multicoliratity_data.xlsx")
```

```r
# 2. Define Variables
y = gdp$Import
x1 = gdp$GDP
x2 = gdp$SF
x3 = gdp$consp
```

```r
# 3. Fit Model
fit = lm(Import ~ GDP + SF + consp, data=gdp)
summary(fit)
```

```r
# 4. Detect Multicollinearity (VIF)
design = cbind(x1, x2, x3)
mctest(fit, type="i", method="VIF")
```

```r
# 5. Remedial Measure: Ridge Regression
# Prepare matrix for glmnet
design_matrix <- as.matrix(design)
```

```r
# Fit Ridge Model (alpha = 0)
ridgem <- cv.glmnet(design_matrix, y, alpha=0)
```

```r
# Optimal Lambda
lambda = ridgem$lambda.min
print(lambda)
```

```r
# Final Coefficients
coef(ridgem, s=lambda)
```
