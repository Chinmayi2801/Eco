```r
# Install and load required packages
install.packages("readxl")
install.packages("lmtest")
library(readxl)
library(lmtest)
```

```r
# 1. Import Data (Ensure 'birdextinct.xlsx' is uploaded to Colab files)
bird <- read_excel("birdextinct.xlsx")
View(bird)
names(bird)
```

```r
# 2. Define Variables
y = bird$"area(km.sq.)"
x1 = bird$"species at risk"
x2 = bird$"# Species extinct"
```

```r
# Calculate proportion
x = x2/x1
```

```r
# 3. Fit the Model
Model = lm(y~x)
summary(Model)
```

```r
# 4. Diagnostic Plots
plot(Model)
```

```r
# 5. Shapiro-Wilk Test (Normality)
shapiro.test(y)
```

```r
# 6. Breusch-Pagan Test (Heteroscedasticity)
bptest(Model)
```

```r
# 7. Remedial Measure: Log Transformation (if needed)
y1 = log(y)
model1.1 = lm(y1~x)
summary(model1.1)
plot(model1.1)
```

```r
# Re-check assumptions on new model
shapiro.test(y1)
bptest(model1.1)
```
