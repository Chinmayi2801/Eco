```r
File Required: Mimosaceae.xlsx
# Install and load packages
install.packages("readxl")
install.packages("lmtest")
library(readxl)
library(lmtest)
```

```r
# 1. Import Data
mimosaceae <- read_excel("Mimosaceae.xlsx")
names(mimosaceae)
```

```r
# 2. Define Variables
y = mimosaceae$"Height (meters)"
x = mimosaceae$"GBH (cm)"
```

```r
# 3. Fit Model
model = lm(y~x)
summary(model)
```

```r
# 4. Detection
plot(model) # Look at Residual vs Fitted and Scale-Location plots
bptest(model) # Breusch-Pagan Test
```

```r
# 5. Normality Check
shapiro.test(y)
```

```r
# 6. Remedial Measure (Log Transformation)
y1 = log(y)
model1.1 = lm(y1~x)
summary(model1.1)
```
